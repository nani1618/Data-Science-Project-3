```python
import pandas as pd
import numpy as np
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
from tabulate import tabulate

df = pd.read_csv("SN_d_tot_V2.0.csv", sep=';')
df['ds'] = pd.to_datetime({'year': df.iloc[:, 0], 'month': df.iloc[:, 1], 'day': df.iloc[:, 2]})
df['y'] = df.iloc[:, 4]
df['y'] = df['y'].replace(-1, np.nan)
df['y'] = df['y'].ffill().bfill()
df['y'] = df['y'] + 1

cutoff_date = df['ds'].max() - pd.DateOffset(years=22)
df_train = df[df['ds'] >= cutoff_date].copy()
```

```python
growth_options = ['linear', 'logistic', 'flat']
periods = [100, 200, 365]

for growth in growth_options:
    if growth == 'logistic':
        df_train['cap'] = df_train['y'].max() * 1.5
    
    model = Prophet(growth=growth, changepoint_prior_scale=0.001, seasonality_prior_scale=25,
                   seasonality_mode='multiplicative', yearly_seasonality=True, weekly_seasonality=False,
                   daily_seasonality=False, n_changepoints=100, interval_width=0.95)
    
    model.add_seasonality(name='solar_cycle', period=365.25 * 11, fourier_order=10, prior_scale=30)
    model.fit(df_train)
    
    for period in periods:
        future = model.make_future_dataframe(periods=period, freq='D')
        if growth == 'logistic':
            future['cap'] = df_train['y'].max() * 1.5
        
        forecast = model.predict(future)
        
        mae = mean_absolute_error(df_train['y'], forecast['yhat'][:len(df_train)])
        rmse = np.sqrt(mean_squared_error(df_train['y'], forecast['yhat'][:len(df_train)]))
        r2 = r2_score(df_train['y'], forecast['yhat'][:len(df_train)])
        
        print(f"\nMetrics for {period} days forecast ({growth} growth):")
        print(f"MAE: {mae:.3f}")
        print(f"RMSE: {rmse:.3f}")
        print(f"RÂ²: {r2:.3f}")
        
        plt.figure(figsize=(15, 8))
        plt.plot(df_train['ds'], df_train['y'], 'b.', alpha=0.3, label='Historical', markersize=1)
        forecast_mask = forecast['ds'] >= df_train['ds'].max()
        plt.plot(forecast.loc[forecast_mask, 'ds'], forecast.loc[forecast_mask, 'yhat'],
                'r-', linewidth=3, label='Forecast')
        plt.fill_between(forecast.loc[forecast_mask, 'ds'],
                        forecast.loc[forecast_mask, 'yhat_lower'],
                        forecast.loc[forecast_mask, 'yhat_upper'],
                        color='r', alpha=0.2)
        plt.title(f'Daily Sunspots Forecast - Next {period} days ({growth} growth)', fontsize=14)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Sunspot Number', fontsize=12)
        plt.legend(fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.xlim(df_train['ds'].max() - pd.DateOffset(years=2), 
                forecast.loc[forecast_mask, 'ds'].max())

        plt.show()
        
        future_predictions = forecast[forecast['ds'] > df_train['ds'].max()]
        pred_table = future_predictions[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
        print("\nFirst 5 predictions:")
        print(tabulate(pred_table, headers='keys', tablefmt='pipe', floatfmt='.2f'))
    
    model.plot_components(forecast)
    plt.show()
```
